## Loading the dataset


```python
import pandas as pd
import numpy as np
import re
```

```python

df=pd.read_csv('https://raw.githubusercontent.com/hantswilliams/HHA_507_2023/main/WK2/data/healthcare_data_cleaning.csv')
df
```

```python
data_types = df.dtypes
print(data_types)
```

## Identify and Handle Missing Value

```python

df.replace(["missing"], np.nan, inplace=True)
```

```python
df.tail(20)
```

```python
# Drop rows with any missing values
df.dropna(inplace=True)
```

```python
# Or if we want to fill missing values for all columns:
df.fillna(0, inplace=True)
df
```

```python
df.describe()
```

What this code does is it checks to see if there are any columns that are missing values. Since there are "missing" for a number of the values that will be converted into NaN and subsequently dropped. The data went from 105,000 rows to 28,666 rows. With the columns remaning the same at 20.

## Clean Column Names

```python
def clean_value(value):
    cleaned_value = re.sub(r'\s+|[^a-zA-Z0-9]', '', value)
    return cleaned_value
```

```python
## clean colomn names
for column in df.columns:
    df[column] = df[column].apply(clean_value)

print(df)
```

I made sure to add import re to my first code cell, to ensure that the built in package would be loaded in and I would be able to "clean the columns." What this code above does is it takes away any white spaces from the column names as well as deletes any special characters.

```python
df.sample(10)
```

## Remove Duplicate Rows and Columns

```python
# Identify and remove duplicate rows
df.drop_duplicates(inplace=True)
df
```

This code removes rows that have the same infromation so as to not skew any data. In this case, rows with no information were already deleted, and subsequently there are no duplicated rows.

```python
# Identify and remove duplicate columns
df2=df.T.drop_duplicates().T
df2
```

This code removes any duplicated columns, and as seen there are none.

## Create New Column

```python
# Define clinically defined thresholds for categorization
age_high = '25'

# Create a new column "age_Status" based on thresholds
df['age_Status'] = df['Patient Age'].apply(lambda x: 'Later Adult' if x >= age_high else 'young adult')

print(df[['Patient Age', 'Gender', 'age_Status']])
```

```python

print(df[['Patient Age', 'Gender', 'age_Status']].head(10))
```

Here, any who is 25 years old or older is considered a Later Adult, and anyone who is younger than 25 is considered a young adult. This column name is referred to as age_Status, and basically just shows if you are a young or later adult.

## Grouby and Pivot Tables

```python
res = df.groupby('State of Residence').agg({
    'Patient Age': 'max'
})
print(res)


```

```python
df['Patient Age'].value_counts()
```

```python
grouped_data = df.groupby('Patient Age').agg({
    'Average BP': 'mean'
})
print(grouped_data)
```

```python
state = df.groupby('Gender').agg({
    'Average BP': 'median'
})
print(state)
```

```python
diagnosis_pivot = df.pivot_table(values='Patient Age', index='Disease Diagnosed', columns='Gender', aggfunc='count')
print(diagnosis_pivot)
```
